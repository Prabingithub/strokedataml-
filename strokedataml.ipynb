```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df, test_df = train_test_split(raw_df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=42)
```

```python
print('train_df.shape:', train_df.shape)
print('val_df.shape:', val_df.shape)
print('test_df.shape:', test_df.shape)
```

```python
# now we are going to identify input and targets columns 
#our inputs are all the columns except id
#our target is the stroke column that we have at last
```

```python
#since id is not our input column we are going to ignore that and leave out the last column that is our output
input_cols=list(train_df.columns)[1:-1]
target_col= 'stroke'
print(input_cols)
```

```python
target_col
```

```python
#now we are goint to create input target and output target for training, testing, validation sets for further processing
train_inputs=train_df[input_cols].copy()
train_targets=train_df[target_col].copy()

#we are using .copy() so the original data set stays the same
```

```python
val_inputs=val_df[input_cols].copy()
val_targets=val_df[target_col].copy()
```

```python
test_inputs=test_df[input_cols].copy()
test_target=test_df[target_col].copy()
```

```python
import numpy as np
```

```python
#we are going to separate categorical and numerical columns so that we can convert categorical data to numeric to train ML model(logistic regression)
numeric_cols=train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_col=train_inputs.select_dtypes('object').columns.tolist()
```

```python
train_inputs[numeric_cols].describe()
```

```python
#lets check how many unique categorical data we have for categorical columns
train_inputs[categorical_col].nunique()
```

```python
#since machine learning cannot woek with missing numeric data we have to fill them all using process called imputation
#let's check the missing values 
raw_df[numeric_cols].isna().sum()
```

```python
train_inputs[numeric_cols].isna().sum()
```

```python
#since we know bmi is the only columns that has no null data, we can replace that using imputation
#we are going to fill the missing data using median because BMI data often contains outliers (e.g., very low or very high BMIs due to extreme health conditions), which can skew the mean. In such cases, the median is generally preferred because it is more robust and provides a more accurate reflection of the typical BMI for most people, without being influenced by the outliers.
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy = 'median')
```

```python
#let us fit imputer to the data
imputer.fit(raw_df[numeric_cols])
```

```python
#lets check what the median for bmi that imputer found and put in the bmi column data 
list(imputer.statistics_)
```

```python
#The missing values in the training, test and validation sets can now be filled in using the transform method of imputer.
train_inputs[numeric_cols]=imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=imputer.transform(test_inputs[numeric_cols])
```

```python
#checking if no null data are replaced 
train_inputs[numeric_cols].isna().sum()
```
